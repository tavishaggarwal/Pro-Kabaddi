## Getting Team Stats
Creating datset containing the details of the team. We are iterating our all the teams and extracting out season wise data for the team.

```python
def getTeamStats():
    url = "https://www.prokabaddi.com/sifeeds/kabaddi/static/json/TEAMNUMBER_team.json"
    df = pd.DataFrame()
    for index in range(1,35):
        r = requests.get(url.replace('TEAMNUMBER', str(index)))
        if r.status_code == 200:
            JSONContent = r.json()
            seasonSize = len(JSONContent['over_all_stats'])
            for season in range(0,seasonSize):
                
                # Ignoring international teams
                if seasonSize and index not in [12, 13, 17, 18, 22]:
                    content = JSONContent['over_all_stats'][season]
                    dict = {'team_name': JSONContent['bio']['team_name'], 'short_name': JSONContent['bio']['short_name'], 'stats_status':
                            content.get('series_name', 'overall'), 'matches': content['matches'], 'wins': content['wins'],
                            'draws': content['draws'], 'losses': content['losses'], 'rank': content['rank'], 'points': content['points'],
                            'score_difference': content['score_difference'], 'all_outs' : content['all_outs'], 
                            'top_raider_id': content['top_raider_id'], 'top_raider_value' :content['top_raider_value'], 
                            'top_defender_id': content['top_defender_id'],
                            'top_defender_value': content['top_defender_value'], 'top_allrounder_id': content['top_allrounder_id'], 
                            'top_all_rounder_value': content['top_all_rounder_value'], 'tackle_points': content['tackle_points'],
                            'capture_points': content['capture_points'], 'tackle_bonus_points' : content['tackle_bonus_points'], 
                            'all_out_points': content['all_out_points'], 'extra_points' :content['extra_points'], 
                            'tackles': content['tackles'], 'tackle_success_rate': content['tackle_success_rate'], 'raids' : content['raids'], 
                            'success_raids': content['success_raids'], 'unsuccess_raids' :content['unsuccess_raids'], 
                            'empty_raid': content['empty_raid'],
                            'success_raid_percent': content['success_raid_percent'], 'green_cards': content['green_cards'], 
                            'yellow_cards': content['yellow_cards'], 'red_cards': content['red_cards'],
                            'raid_bonus_points': content['raid_bonus_points'], 'success_tackles' : content['success_tackles'], 
                            'defence_touch_points': content['defence_touch_points'], 'unsuccesful_tackles' :content['defending_bonus_points'], 
                            'raid_touch_points': content['raid_touch_points'], 'unsuccesful_tackles': content['unsuccesful_tackles'], 
                            'total_raid_points' : content['total_raid_points'],
                            'total_defending_points': content['total_defending_points'], 'super_tackles' :content['super_tackles'], 
                            'super_raids': content['super_raids'],
                            'highest_score_team_name': content['highest_score']['team_name'], 
                            'highest_score': content['highest_score']['highest_score'], 
                            'max_win_margin_team1_name' : content['max_win_margin']['team1_name'],
                            'max_win_margin_team2_name': content['max_win_margin']['team2_name'], 
                            'max_win_margin_team1_score' :content['max_win_margin']['team1_score'], 
                            'max_win_margin_team2_score': content['max_win_margin']['team2_score'],
                            'max_win_margin': content['max_win_margin']['max_win_margin'],
                            'match_date': content['max_win_margin']['match_date']
                           }
                    df = df.append(dict, ignore_index=True)
    return df
```

```python
seasons = [
    'https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_1_96_stats.json',
    'https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_2_96_stats.json',
    'https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_3_96_stats.json',
    'https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_4_96_stats.json',
    'https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_8_96_stats.json',
    'https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_26_96_stats.json',
    'https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_49_96_stats.json',
    'https://www.prokabaddi.com/sifeeds/kabaddi/static/json/1_0_96_stats.json'
]

season_list = ['Season 1', 'Season 2', 'Season 3', 'Season 4', 'Season 5', 'Season 6',  'Season 7', 'Overall']

def getTotalPointData():
    df = pd.DataFrame()
    for season in range(0,len(seasons)):
        r = requests.get(seasons[season])
        if r.status_code == 200:
            JSONContent = r.json()
            for index in range(0,len(JSONContent['data'])):
                dict = {
                        'season': season_list[season],
                        'match_played': JSONContent['data'][index]['match_played'],
                        'rank': JSONContent['data'][index]['rank'],
                        'team_id': JSONContent['data'][index]['team_id'],
                        'team_name': JSONContent['data'][index]['team_name'],
                        'total_points': JSONContent['data'][index]['value'],
                       }
                df = df.append(dict, ignore_index=True)
    return df
```

## Getting Match Stats
Getting details of the match played in the each seasons. We are iterating our all the matches played in pro kabbadi till date across all the seasons.

```python
def getMatchStats():
    url = "https://www.prokabaddi.com/sifeeds/kabaddi/live/json/MACTHID_match.json"
    df = pd.DataFrame()
    for index in range(1,2000):
        r = requests.get(url.replace('MACTHID', str(index)))
        if r.status_code == 200:
            JSONContent = r.json()
#             print(JSONContent)
            dict = {'SeasonId': JSONContent['match_detail']['series']['name'], 
                    'MatchId': JSONContent['match_detail']['match_number'],
                    'home_team_name': JSONContent['teams'].get('home_team_name',None), 
                    'Team1': JSONContent['teams']['team'][0]['name'],
                    'Team2': JSONContent['teams']['team'][1]['name'], 
                    'Team1Score': JSONContent['teams']['team'][0]['score'],
                    'Team2Score': JSONContent['teams']['team'][1]['score'],
                    'Team2_all_outs_conceded': JSONContent['teams']['team'][0]['stats']['all_outs'],
                    'Team1_all_outs_conceded': JSONContent['teams']['team'][1]['stats']['all_outs']
                   }
            
            if JSONContent['match_detail']['result']:
                dict['MatchResult'] = JSONContent['match_detail']['result']['value']
            else:
                dict['MatchResult'] = None
            df = df.append(dict, ignore_index=True)
    return df
```

## Get Stats
Get overall Stats of the pro kabaddi seasons

```python
def getAllstats(url):
    df = pd.DataFrame()
    url = url
    r = requests.get(url)
    if r.status_code == 200:
        JSONContent = r.json()

        for index in range(1,len(JSONContent['all_stats'])):
            for j in range(1,len(JSONContent['all_stats'][index]['data'])):
                dict = {
                       'match_played': JSONContent['all_stats'][index]['data'][j]['match_played'], 
                       'player': JSONContent['all_stats'][index]['data'][j]['player'],
                        'player_id': JSONContent['all_stats'][index]['data'][j]['player_id'],
                        'player_name': JSONContent['all_stats'][index]['data'][j]['player_name'],
                        'position_id': JSONContent['all_stats'][index]['data'][j]['position_id'],
                        'position_name': JSONContent['all_stats'][index]['data'][j]['position_name'],
                        'rank': JSONContent['all_stats'][index]['data'][j]['rank'],
                       'tackles': JSONContent['all_stats'][index]['data'][j]['tackles'],
                        'team': JSONContent['all_stats'][index]['data'][j]['team'],
                        'team_full_name': JSONContent['all_stats'][index]['data'][j]['team_full_name'],
                        'team_id': JSONContent['all_stats'][index]['data'][j]['team_id'],
                         'total_defending_points': JSONContent['all_stats'][index]['data'][j]['total_defending_points'],
                        'total_points': JSONContent['all_stats'][index]['data'][j]['value'],
                    'minimum_criteria': JSONContent['all_stats'][index]['minimum_criteria'],
                    'stat': JSONContent['all_stats'][index]['stat'],
                    'stat_id': JSONContent['all_stats'][index]['stat_id'],
                    'stat_type': JSONContent['all_stats'][index]['stat_type']
                    }
                df = df.append(dict, ignore_index=True)
    return df
```
